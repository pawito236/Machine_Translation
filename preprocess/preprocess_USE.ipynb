###Data V1

'm
Remove { 0 }. ".....",

```python
def cleansing_v1(df):

  # cut "\n"
  df['en'] = df['en'].str.replace("\n","")
  df['th'] = df['th'].str.replace("\n","")

  # cut "�"
  df['en'] = df['en'].str.replace("�","'")

  # cut "--"
  df['th'] = df['th'].str.replace("--","")
  df['en'] = df['en'].str.replace("--","")

  # fix "i m"
  df['en'] = df['en'].str.replace("i ' m","i'm")

  # remove row of {0}
  list_del = df[df['en'].str.contains("{ 0 }")].index
  df.drop(index=list_del, inplace = True)

  # Replace "can" to "Can" ,what, where ,when, why, who , how , 
  def change_Bletter_q(df, word_start, word_end, word_start_new):
    index = df[(df.en.str.startswith(word_start) == True) & (df.en.str.endswith(word_end)== True)].index
    df['en'][index] = df[(df.en.str.endswith(word_end) == True) & (df.en.str.startswith(word_start) == True)]['en'].str.replace(word_start,word_start_new)
    return df
  change_Bletter_q(df, "can", "?", "Can")
  change_Bletter_q(df, "what", "?", "What")
  change_Bletter_q(df, "where", "?", "Where")
  change_Bletter_q(df, "when", "?", "When")
  change_Bletter_q(df, "why", "?", "Why")
  change_Bletter_q(df, "who", "?", "Who")
  change_Bletter_q(df, "how", "?", "How")
  change_Bletter_q(df, "could", "?", "Could")

  df['en'] = df['en'].str.replace("\\......","")
  df['en'] = df['en'].str.replace("\\.....","")
  df['en'] = df['en'].str.replace("\\....","")
  df['en'] = df['en'].str.replace("\\...","")
  df['en'] = df['en'].str.replace("\\..","")
  
  return df
```

```python
def cleansing_v3(df):
  # cut "\n"
  df['en'] = df['en'].str.replace("\n","")
  df['th'] = df['th'].str.replace("\n","")

  # &apos;s
  df['en'] = df['en'].str.replace("&apos;","'")

  # Not endswith fullstop and not endswith "?"  -> Keep index in E_idx
  E_idx = df[(df['en'].str.endswith('.') == False)&(df['en'].str.endswith('?') == False)]['en'].index
  # add fullstop to all non question by E_idx
  df['en'][E_idx] = df[(df['en'].str.endswith('.') == False)&(df['en'].str.endswith('?') == False)]['en'] + " ."
  # th endswith Fullstop
  E_idx = df[df['th'].str.endswith('.') == True].index
  # add fullstop to all non question by E_idx
  df['th'][E_idx] = df[(df['th'].str.endswith('.') == True)]['th'].str.replace(".","")
  E_idx = df[(df['th'].str.endswith('?') == False) & (df['en'].str.endswith('?') == True)].index
  # add question mark to all non conversation by E_idx
  df['th'][E_idx] = df[(df['th'].str.endswith('?') == False) & (df['en'].str.endswith('?') == True)]['th'] + " ?"
  E_idx = df[(df['th'].str.endswith('?') == True) & (df['en'].str.endswith('?') == False)].index
  # add ? to all non question by E_idx 
  # ปย en ที่ไม่ลงท้ายด้วย "." แต่ไทยเป็น คำถาม
  df['en'][E_idx] = df[(df['th'].str.endswith('?') == True) & (df['en'].str.endswith('?') == False)]['en'].str.replace(".","?")

  # เว้น . ? ตอยจบ ปย.
  # ---------------> EN
  E_idx = df[(df['en'].str.endswith(' .') == False) & (df['en'].str.endswith('.') == True)].index
  df['en'][E_idx] = (df[(df['en'].str.endswith(' .') == False) & (df['en'].str.endswith('.') == True)]['en'] + "Sleepyyyy").str.replace(".Sleepyyyy"," .")
  E_idx = df[(df['en'].str.endswith(' ?') == False) & (df['en'].str.endswith('?') == True)].index
  df['en'][E_idx] = (df[(df['en'].str.endswith(' ?') == False) & (df['en'].str.endswith('?') == True)]['en'] + "Sleepyyyy").str.replace(".Sleepyyyy"," ?")
  # -----------------> TH
  E_idx = df[(df['th'].str.endswith(' ?') == False) & (df['th'].str.endswith('?') == True)].index
  df['th'][E_idx] = (df[(df['th'].str.endswith(' ?') == False) & (df['th'].str.endswith('?') == True)]['th'] + "Sleepyyyy").str.replace(".Sleepyyyy"," ?")

  # sth -> something, sb - > somebody
  df['en'] = df['en'].str.replace(' sb . '," somebody ")
  df['en'] = df['en'].str.replace('sb . ',"somebody ")
  df['en'] = df['en'].str.replace(' sb '," somebody ")
  df['en'] = df['en'].str.replace('sb ',"somebody ")
  df['en'] = df['en'].str.replace(' sth '," something ")
  df['en'] = df['en'].str.replace('sth ',"something ")
  df['th'] = df['th'].str.replace("s b","บางคน")
  df['th'] = df['th'].str.replace("st h","บางอย่าง")

  # Remove **** when th has * but en not
  E_idx = df[(df['en'].str.contains('\\*') == True) & (df['th'].str.contains('\\*') == False)].index
  df['en'][E_idx] = df[(df['en'].str.contains('\\*') == True) & (df['th'].str.contains('\\*') == False)]['en'].str.replace("*","")

  # cut "�"
  df['en'] = df['en'].str.replace("�","'")

  # cut "--"
  df['th'] = df['th'].str.replace("--","")
  df['en'] = df['en'].str.replace("--","")

  # fix "i m"
  df['en'] = df['en'].str.replace("i ' m","i'm")

  # remove row of {0}
  list_del = df[df['en'].str.contains("{ 0 }")].index
  df.drop(index=list_del, inplace = True)

  # Replace "can" to "Can" ,what, where ,when, why, who , how , 
  def change_Bletter_q(df, word_start, word_end, word_start_new):
    index = df[(df.en.str.startswith(word_start) == True) & (df.en.str.endswith(word_end)== True)].index
    df['en'][index] = df[(df.en.str.endswith(word_end) == True) & (df.en.str.startswith(word_start) == True)]['en'].str.replace(word_start,word_start_new)
    return df
  change_Bletter_q(df, "can", "?", "Can")
  change_Bletter_q(df, "could", "?", "Could")
  change_Bletter_q(df, "may", "?", "May")
  change_Bletter_q(df, "might", "?", "Might")
  change_Bletter_q(df, "can", "?", "Can")
  change_Bletter_q(df, "what", "?", "What")
  change_Bletter_q(df, "where", "?", "Where")
  change_Bletter_q(df, "when", "?", "When")
  change_Bletter_q(df, "why", "?", "Why")
  change_Bletter_q(df, "who", "?", "Who")
  change_Bletter_q(df, "how", "?", "How")
  change_Bletter_q(df, "could", "?", "Could")

  # fill Q in th
  def fill_q(df, end_word, new_end_word):
    idx = df[df.th.str.endswith(end_word)== True].index
    df['th'][idx] = df[(df.th.str.endswith(end_word) == True)]['th'].str.replace(end_word,new_end_word)
    return df
  fill_q(df, 'หรือ ไม่', 'หรือ ไม่ ?')
  fill_q(df, 'อย่าง ไร', 'อย่าง ไร ?')
  fill_q(df,"หรือ เปล่า", "หรือ เปล่า ?")

  df['en'] = df['en'].str.replace("\\......","")
  df['en'] = df['en'].str.replace("\\.....","")
  df['en'] = df['en'].str.replace("\\....","")
  df['en'] = df['en'].str.replace("\\...","")
  df['en'] = df['en'].str.replace("\\..","")
  
  # add for v3
  df['en'] = df['en'].str.replace('\ % s',"")
  df['en'] = df['en'].str.replace(': % s',"")
  df['en'] = df['en'].str.replace('( % s )',"")
  df['en'] = df['en'].str.replace('% u',"")
  df['en'] = df['en'].str.replace('% d /',"")
  df['en'] = df['en'].str.replace('\ % d',"")
  df['en'] = df['en'].str.replace('% d',"")
  df['en'] = df['en'].str.replace('% m / % d / % y',"")
  df['en'] = df['en'].str.replace('% m / % d / % E y',"")

  df['th'] = df['th'].str.replace('\ % s',"")
  df['th'] = df['th'].str.replace(': % s',"")
  df['th'] = df['th'].str.replace('( % s )',"")
  df['th'] = df['th'].str.replace('% u',"")
  df['th'] = df['th'].str.replace('% d /',"")
  df['th'] = df['th'].str.replace('\ % d',"")
  df['th'] = df['th'].str.replace('% d',"")
  df['th'] = df['th'].str.replace('% m / % d / % y',"")
  df['th'] = df['th'].str.replace('% m / % d / % E y',"")


  return df
```

```python
def get_similar_score(lang1: list, lang2:list , batch_size:int):
  '''
  get_similar_score is a function to get similarity score of sentence pair


  Input: 
   lang1: list of sentence of lang1 
   lang2: list of sentence of lang2 
   batch_size: batch_size

   ** As sometime data is too big, too do on one go
      the we need to do it as a batch
  
  outout: 
   scores: list of score
  '''

  scores =[]

  if len(lang1)%batch_size != 0:
    num_of_batch = int(len(lang1)/batch_size)+1
  else:
    num_of_batch = int(len(lang1)/batch_size)
  
  print(num_of_batch)

  for i in tqdm(range(num_of_batch)):
    start = i*batch_size
    end = start+batch_size
    if i <= num_of_batch:

      lang1_temp = lang1[start:end]
      lang2_temp = lang2[start:end]
      
      lang1_embedding = embed(lang1_temp)
      lang2_embedding = embed(lang2_temp)
      distance_matrix = tf.matmul(lang1_embedding, lang2_embedding, transpose_b=True).numpy()

      for j in range(len(distance_matrix)):
        scores.append(distance_matrix[j][j])

  return scores
```
